```python
categorical_columns = ['sex', 'smoker', 'day', 'time']

encoder = OrdinalEncoder(categories=[
    ['Male', 'Female'],
    ['No', 'Yes'],
    ['Sat', 'Sun', 'Thur', 'Fri'],
    ['Dinner', 'Lunch']
])
```

```python
df_encoded = df.copy()
df_encoded[categorical_columns] = encoder.fit_transform(df[categorical_columns])

print(df_encoded)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse_output=False)
```

```python
encoded_array = ohe.fit_transform(df[categorical_columns])
```

```python
encoded_df = pd.DataFrame(
    encoded_array, columns=ohe.get_feature_names_out(categorical_columns))

df_final = pd.concat([df.drop(columns=categorical_columns), encoded_df], axis=1)

df_final
```

```python
from sklearn.compose import ColumnTransformer
```

```python
numerical_columns = ['total_bill', 'tip', 'size']
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(sparse_output=False, drop=None), categorical_columns),
        ('passthrough', 'passthrough', numerical_columns),
    ]
)
```

```python
transformed_data = preprocessor.fit_transform(df)
```

```python
onehot_columns = preprocessor.named_transformers_['onehot'].get_feature_names_out(categorical_columns)
all_columns = list(onehot_columns) + numerical_columns
```

```python
df_transformed = pd.DataFrame(transformed_data, columns=all_columns)
```
