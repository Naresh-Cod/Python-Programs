```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head()
```

```python
df = df.rename(columns={'per capita income (US$)':"cap_income"})
```

```python
new_df = df.drop('cap_income', axis='columns')
new_df.head()
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per capita income (US$)')
plt.scatter(df.year,df.cap_income,color='red',marker='+')
```

```python
df.head()
```

```python
model = linear_model.LinearRegression()
model.fit(new_df, df.cap_income)
```

```python
model.predict([[2020]])
```

```python
model.coef_
```

```python
model.intercept_
```
