```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = {'Mileage':[69000, 35000, 57000, 22500, 46000, 59000, 52000, 72000, 91000, 67000, 83000, 79000, 59000, 58780, 82450, 25400, 28000, 69000, 87600, 52000],
        'Age(yrs)':[6, 3, 5, 2, 4, 5, 5, 6, 8, 6, 7, 7, 5, 4, 7, 3, 2, 5, 8, 5],
        'Sell Price($)':[18000, 34000, 26100, 40000, 31500, 26750, 32000, 19300, 12000, 22000, 18700, 19500, 26000, 27500, 19400, 35000, 35500, 19700, 12800, 28200]}
df = pd.DataFrame(data)
df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.xlabel('Mileage')
plt.ylabel('price')
plt.scatter(df.Mileage, df['Sell Price($)'])
```

```python
plt.xlabel('Age(yrs)')
plt.ylabel('price')
plt.scatter(df['Age(yrs)'], df['Sell Price($)'])
```

```python
x = df[['Mileage', 'Age(yrs)']]
y = df[['Sell Price($)']]
```

```python
len(x), len(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)
```

```python
(len(x_train), len(x_test)), (len(y_train), len(y_test))
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.predict([[67000, 6]])
```

```python
x_test
```

```python
model.predict(x_test)
```

```python
y_test
```

```python
model.score(x_test, y_test)
```
